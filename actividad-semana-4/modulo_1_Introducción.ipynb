### Agregar encabezados

Echa un vistazo a nuestro conjunto de datos. Pandas establece automáticamente el encabezado con un número entero a partir de 0.
Para describir mejor nuestros datos, podemos introducir un encabezado.
Por lo tanto, tenemos que agregar encabezados manualmente.
Primero, creamos una lista de "encabezados" que incluyen todos los nombres de las columnas en orden. Luego, usamos dataframe.columns = headers para reemplazar los encabezados con la lista que creamos.

```python
headers =  ['simbolización', 'pérdidas normalizadas', 'marca', 'tipo de combustible', 'aspiración', 'número de puertas', 
            'estilo de la carrocería', 'ruedas motrices', 'ubicación del motor' , 'distancia entre ejes', 'longitud', 
            'ancho', 'altura', 'peso en vacío', 'tipo de motor', 'número de cilindros', 'tamaño del motor', 'sistema de combustible', 
            'diámetro interior', 'carrera', 'relación de compresión', 'caballos de fuerza', 'rpm pico', 'mpg en ciudad', 
            'mpg en carretera', 'precio']

print("Encabezados\n", headers)
```

```python
df.columns = headers
df.head(10)
```

Necesitamos reemplazar el símbolo "?" con NaN para que dropna() pueda eliminar los valores faltantes:

## Manejar valores faltantes

¿Cómo trabajar con datos faltantes?

Pasos para trabajar con datos faltantes:

* Identificar datos faltantes
* Tratar con datos faltantes
* Formato de datos correcto

### Identificar valores faltantes
Convertir "?" a NaN
En el conjunto de datos del automóvil, los datos faltantes vienen con el signo de interrogación "?". Reemplazamos "?" con NaN (No es un número).

```python
import numpy as np
import pandas as pd

df = pd.read_csv(path, header=None)

df.replace("?", np.nan, inplace = True)
df.head(5)
```

### Evaluación de datos faltantes
Los valores que faltan se convierten de forma predeterminada. Existen dos métodos para detectar datos faltantes:

* .isnull()
* .notnull()

La salida es un valor booleano que indica si el valor que se pasa al argumento indica que faltan de datos. "Verdadero" significa que el valor es un valor perdido, mientras que "Falso" significa que el valor no es un valor perdido.

```python
df.columns = headers
missing_data = df.isnull()
missing_data.head(5)
```

Cuente los valores faltantes en cada columna

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

####Tratar con datos faltantes
¿Cómo lidiar con los datos faltantes?
#####**Borrar los datos**
* Borrar toda la fila
* Borrar toda la columna

#####**Reemplazar datos**
* Reemplázarlos por la media
* Reemplazarlos por la frecuencia
* Reemplázalos en función de otras funciones

Calcule el valor medio de la columna "pérdidas normalizadas"

```python
avg_norm_loss = df["pérdidas normalizadas"].astype("float").mean(axis=0)
print("Media de las pérdidas normalizadas: ", avg_norm_loss)
```

Reemplace "NaN" con el valor medio en la columna "pérdidas normalizadas"

```python
df["pérdidas normalizadas"].replace(np.nan, avg_norm_loss, inplace=True)
```
