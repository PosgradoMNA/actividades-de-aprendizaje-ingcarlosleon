##Identificar y tratar los valores que faltan
###Identificar los valores que faltan
####Convertir "?" en NaN
En el conjunto de datos del coche, los datos que faltan vienen con el signo de interrogación "?". Sustituimos "?" por NaN (Not a Number), el marcador de valores perdidos por defecto de Python, por razones de velocidad de cálculo y comodidad. Aquí utilizamos la función

````python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
````

###Evaluación de los datos que faltan

Los valores perdidos se convierten por defecto. Utilizamos las siguientes funciones para identificar estos valores perdidos. Hay dos métodos para detectar los datos que faltan:

*   .isnull()
*   .notnull()

La salida es un valor booleano que indica si el valor que se pasa en el argumento es en realidad un dato que falta.

````python
missing_data = df.isnull()
missing_data.head(5)
````

"Verdadero" significa que el valor es un valor perdido mientras que "Falso" significa que el valor no es un valor perdido.

##Contar los valores que faltan en cada columna

Utilizando un bucle for en Python, podemos averiguar rápidamente el número de valores que faltan en cada columna. Como se ha mencionado anteriormente, "True" representa un valor que falta y "False" significa que el valor está presente en el conjunto de datos. En el cuerpo del bucle for el método ".value_counts()" cuenta el número de valores "True".

````python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
````

##Normalización de datos
**¿Por qué la normalización?**

La normalización es el proceso de transformar los valores de varias variables en un rango similar. Las normalizaciones típicas incluyen el escalado de la variable para que la media de la variable sea 0, el escalado de la variable para que la varianza sea 1, o el escalado de la variable para que los valores de la variable oscilen entre 0 y 1.

**Ejemplo**

Para demostrar la normalización, digamos que queremos escalar las columnas "longitud", "anchura" y "altura".

**Objetivo:** queremos normalizar esas variables para que su valor vaya de 0 a 1

**Método:** sustituir el valor original por (valor original)/(valor máximo)

````python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
````

##Ejemplo de binning de datos en Pandas


Convertir los datos al formato correcto:


````python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
````

Vamos a trazar el histograma de la potencia para ver cómo es la distribución de la potencia.

````python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
````

Queremos 3 bins de igual tamaño de ancho de banda, así que usamos la función linspace(valor_inicial, valor_final, números_generados) de numpy.

Como queremos incluir el valor mínimo de los caballos de fuerza, queremos establecer start_value = min(df["horsepower"]).

Como queremos incluir el valor máximo de caballos de fuerza, queremos establecer end_value = max(df["horsepower"]).

Dado que estamos construyendo 3 bins de igual longitud, debe haber 4 divisores, por lo que numbers_generated = 4.

Construimos una matriz de recipientes con un valor mínimo a un valor máximo utilizando el ancho de banda calculado anteriormente. Los valores determinarán cuándo termina un bin y cuándo empieza otro.

````python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
````

Fijamos los nombres de los grupos:


````python
group_names = ['Low', 'Medium', 'High']
````

Aplicamos la función "cortar" para determinar a qué pertenece cada valor de df['horsepower'].

````python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
````

Veamos el número de vehículos en cada contenedor:


````python
df["horsepower-binned"].value_counts()
````

Vamos a graficar la distribución de cada bin:


````python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
````

Observe el marco de datos anterior con atención. Encontrará que la última columna proporciona los intervalos de "caballos de fuerza" basados en 3 categorías ("Bajo", "Medio" y "Alto").

Hemos conseguido reducir los intervalos de 59 a 3.

##Variable indicadora (o variable ficticia)
**¿Qué es una variable indicadora?**
Una variable indicadora (o variable ficticia) es una variable numérica utilizada para etiquetar categorías. Se denominan "ficticias" porque los números en sí no tienen un significado inherente.

**¿Por qué utilizamos variables indicadoras?**

Utilizamos variables indicadoras para poder utilizar variables categóricas para el análisis de regresión en los módulos posteriores.

**Ejemplo:**
Vemos que la columna "tipo de combustible" tiene dos valores únicos: "gasolina" o "diesel". La regresión no entiende de palabras, sólo de números. Para utilizar este atributo en el análisis de regresión, convertimos "tipo de combustible" en variables indicadoras.

Utilizaremos el método de pandas 'get_dummies' para asignar valores numéricos a las diferentes categorías de tipo de combustible.

````python
df.columns
````

Obtenga las variables indicadoras y asígnelas al marco de datos "dummy_variable_1":

````python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
````

Cambie los nombres de las columnas para mayor claridad:


````python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
````

En el marco de datos, la columna "fuel-typ" tiene valores para "gas" y "diesel" como 0s y 1s ahora.

````python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
````

````python
df.head()
````

Las dos últimas columnas son ahora la representación de la variable indicadora del tipo de combustible. Ahora son todos 0s y 1s.
